```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Read the data from the text file
data = np.loadtxt("Q_4_Exponential_Numbers.txt")

# Generate points for the exponential PDF
x = np.linspace(0, 5, 1000)
pdf = expon.pdf(x, scale=0.5)  # Exponential PDF with mean 0.5

# Plot a density histogram of the generated random numbers
plt.hist(data, bins=30, density=True, color='skyblue', edgecolor='black', alpha=0.7, label='Density Histogram')

# Plot the exponential PDF for comparison
plt.plot(x, pdf, color='red', linestyle='dashed', linewidth=2, label='Exponential PDF')

# Add labels and title
plt.title('Density Histogram and Exponential PDF')
plt.xlabel('Value')
plt.ylabel('Density')

# Add legend
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```
