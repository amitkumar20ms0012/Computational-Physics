```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Gaussian probability density function (PDF)
def gaussian_pdf(x):
    return 1 / np.sqrt(2 * np.pi) * np.exp(-0.5 * x**2)

# Number of random numbers to generate
n = 10000

# Generate random numbers using Box-Muller method
u1 = np.random.rand(n)
u2 = np.random.rand(n)
z1 = np.sqrt(-2 * np.log(u1)) * np.cos(2 * np.pi * u2)
z2 = np.sqrt(-2 * np.log(u1)) * np.sin(2 * np.pi * u2)

# Concatenate z1 and z2 to get the desired number of samples
z = np.concatenate((z1, z2))

# Plotting
plt.figure(figsize=(8, 6))

# Plot density histogram
plt.hist(z, bins=50, density=True, color='skyblue', edgecolor='black', alpha=0.7, label='Density Histogram')

# Plot true Gaussian PDF
x = np.linspace(-5, 5, 1000)
plt.plot(x, gaussian_pdf(x), color='red', linestyle='dashed', linewidth=2, label='Gaussian PDF')

# Add labels and title
plt.title('Density Histogram and Gaussian PDF')
plt.xlabel('Value')
plt.ylabel('Density')

# Add legend
plt.legend()

# Show plot
plt.grid(True)
plt.show()

# The Box-Muller method generates pairs of random numbers (z1 and z2) that are independently and identically distributed (iid) samples from a standard normal distribution (mean 0, variance 1). By combining these two sets of numbers into a single array, we effectively double the number of samples while still maintaining the desired distribution properties.

# This allows us to generate 10,000 random numbers in total, which is what the user requested.
```
