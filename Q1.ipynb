```python
import matplotlib.pyplot as plt
import time as t
import numpy as np

# Parameters for the LCG
a = 1103515245 # ANSI C standard
c = 12345
m = 2**31  # Corrected modulus to integer

# Initial seed
seed = 1.1

# Array to hold random numbers
n = 10000
x_arr = np.zeros(n, dtype=np.float64)
x = seed

# Start timing
start_time = t.time()

# Generate 10,000 random numbers
for i in range(n):
    x = (a * x + c) % m
    x_arr[i] = x

# End timing
elapsed_time = t.time() - start_time

# Normalize the numbers to [0, 1]
x_arr = x_arr / m
print("Time Taken = ", elapsed_time)

# Plotting the density histogram
plt.plot(np.linspace(0, 1, 2), np.ones(2), color='black', label="True Uniform Distribution")
plt.hist(x_arr, bins=50, density=True, facecolor='steelblue', label="Distribution Density Histogram", histtype='stepfilled')
plt.title("Random Numbers using Linear Congruential Generator", size=13)
plt.xlabel("x")
plt.ylabel("p(x)")
plt.grid()
plt.legend()
plt.show()
```
