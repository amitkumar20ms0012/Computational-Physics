```python
# Importing required libraries
import numpy as np
import matplotlib.pyplot as plt
import time as t

# Starting timing
t_val = np.float64(t.time())

# Generating 10,000 random numbers from a uniform distribution [0, 1]
x = np.random.rand(10000)

# Ending timing
t_val = np.float64(t.time()) - t_val
print("Time Taken = ", t_val)

# Plotting the true uniform distribution line
plt.plot(np.linspace(0, 1, 2), np.ones(2), color='black', label="Uniform Distribution")

# Plotting the density histogram of the generated random numbers
plt.hist(x, 20, density=True, facecolor='skyblue', label="Random Numbers Histogram", histtype='stepfilled')

# Adding title and labels to the plot
plt.title("Random numbers using Numpy", size=13)
plt.xlabel("x")
plt.ylabel("p(x)")

# Adding grid and legend to the plot
plt.grid()
plt.legend()

# Displaying the plot
plt.show()
```
